## Helpers for visualization

````python
# =========================
# VALIDATION + VISUALIZATION
# =========================
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


def list_npz_files(root: Path, max_files=5):
    files = list(root.rglob("*.npz"))
    files = [f for f in files if f.parent.name=='all'] or files
    return files[:max_files]

def show_triplet(img3, title=""):
    fig, axes = plt.subplots(1,3, figsize=(9,3))
    names = ["science","template","difference"]
    for i,ax in enumerate(axes):
        ar = img3[i]
        finite = np.isfinite(ar)
        if finite.any():
            vmin, vmax = np.percentile(ar[finite], [5, 99])
            if vmin == vmax: vmax = vmin + 1e-6
            ax.imshow(ar, origin='lower', vmin=vmin, vmax=vmax)
        else:
            ax.imshow(np.zeros_like(ar))
        ax.set_title(names[i])
        ax.axis('off')
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

def _pick_meta_columns(meta_cols):
    """Choose a readable subset of meta columns that are likely to exist."""
    preferred = [
        # alert quality/geometry
        'rb','sgscore1','sgscore2','distpsnr1','distpsnr2','sharpnr','scorr','fwhm',
        'sigmapsf','diffmaglim','chinr','classtar',
        # context extras we added
        'days_to_peak','days_since_peak','age_sum_days',
        'peakmag_so_far','maxmag_so_far','max_over_peak_mag',
        'n_photometry_total','n_photometry_fid_1','n_photometry_fid_2','n_photometry_fid_3'
    ]
    return [c for c in preferred if c in meta_cols]

def _plot_alert_dt_hist(prov_df):
    if prov_df is None or 'alert_dt_days' not in prov_df.columns:
        return
    plt.figure(figsize=(5,3))
    plt.hist(prov_df['alert_dt_days'].astype(float), bins=20, alpha=0.8)
    plt.xlabel("Δt = |JD_alert - JD_event| (days)")
    plt.ylabel("# events")
    ttl = "Alert–Event Time Offsets"
    if 'select_policy' in prov_df.columns:
        counts = prov_df['select_policy'].value_counts().to_dict()
        ttl += "  " + "  ".join([f"{k}:{v}" for k,v in counts.items()])
    plt.title(ttl)
    plt.tight_layout()
    plt.show()

def _plot_lc_with_spectrum_marker(event_data, event_cols, spectrum_dt=None):
    if 'dt' not in event_cols or 'band_id' not in event_cols or 'logflux' not in event_cols:
        return
    dt_idx  = event_cols.index('dt')
    band_idx= event_cols.index('band_id')
    lf_idx  = event_cols.index('logflux')
    dt = event_data[:, dt_idx]
    band = event_data[:, band_idx].astype(int)
    lf = event_data[:, lf_idx]
    plt.figure(figsize=(6,4))
    for b in np.unique(band):
        m = band==b
        plt.plot(dt[m], lf[m], '.', label=f'band {b}', alpha=0.8)
    if spectrum_dt is not None and np.isfinite(spectrum_dt):
        plt.axvline(float(spectrum_dt), linestyle='--', linewidth=2, label=f'spectrum @ {float(spectrum_dt):.3f} d')
    plt.xlabel("dt (days)"); plt.ylabel("log10(flux)"); plt.legend(); plt.title("Merged LC (events)")
    plt.tight_layout()
    plt.show()

def _plot_spectrum(wave, flux, title="Spectrum"):
    plt.figure(figsize=(7,3.5))
    if wave is None:
        # assume pixel index if wavelengths missing
        x = np.arange(len(flux))
        plt.plot(x, flux, lw=1)
        plt.xlabel("pixel")
    else:
        plt.plot(wave, flux, lw=1)
        plt.xlabel("wavelength")
    plt.ylabel("normalized flux")
    plt.title(title)
    plt.tight_layout()
    plt.show()

def validate_npz(npz_path: Path, match_tol_days=cfg.alert_tol_days, n_show=2):
    print(f"\n==== {npz_path} ====")
    z = np.load(npz_path, allow_pickle=True)

    # core tensors
    images     = z['images']           # (T, 3, H, W)
    event_data = z['event_data']       # (T, Fe)
    event_cols = [c for c in z['event_columns']]
    meta_data  = z['meta_data']        # (T, Fm)
    meta_cols  = [c for c in z['meta_columns']]
    label      = int(z['label'])
    label_str  = z['label_str'][()] if 'label_str' in z.files else str(label)
    print(f"label: {label} ({label_str}), T={len(images)}, Fe={len(event_cols)}, Fm={len(meta_cols)}")

    # spectra (optional)
    has_spec = int(np.array(z.get('has_spectrum', 0)).item()) if 'has_spectrum' in z.files else int('spectrum' in z.files)
    spectrum = None; spectrum_w = None; spectrum_dt = None; spectrum_jd = None
    if 'spectrum' in z.files:
        spectrum = z['spectrum']
        if spectrum.ndim == 2 and spectrum.shape[0] == 1:
            spectrum = spectrum[0]
    if 'spectrum_wavelength' in z.files:
        spectrum_w = z['spectrum_wavelength']
    if 'spectrum_dt' in z.files:
        spectrum_dt = float(np.array(z['spectrum_dt']).item())
    if 'spectrum_jd' in z.files:
        spectrum_jd = float(np.array(z['spectrum_jd']).item())
    if has_spec:
        print(f"spectrum present: yes | dt={spectrum_dt} d (rel. to first phot. MJD){' | jd='+str(spectrum_jd) if spectrum_jd is not None else ''}")
    else:
        print("spectrum present: no")

    # provenance checks
    prov_df = None
    if 'provenance' in z.files:
        prov_df = pd.DataFrame(list(z['provenance']))
        if 'alert_dt_days' in prov_df.columns:
            bad = prov_df[prov_df['alert_dt_days'].astype(float) > match_tol_days]
            print(f"matched events outside tol: {len(bad)} / {len(prov_df)} (tol={match_tol_days} d)")
        if 'select_policy' in prov_df.columns:
            print("select_policy counts:", prov_df['select_policy'].value_counts().to_dict())
        _plot_alert_dt_hist(prov_df)
    else:
        print("No provenance in file.")

    # leakage sanity for peak metrics if jd present
    ok_leak = 'days_since_peak' in meta_cols and 'days_to_peak' in meta_cols
    if ok_leak and 'jd' in z.files and 'logflux' in event_cols:
        jd = z['jd'].astype(float)
        lf_idx = event_cols.index('logflux')
        flux = 10.0 ** (event_data[:, lf_idx])
        ds_idx = meta_cols.index('days_since_peak')
        dtp_idx= meta_cols.index('days_to_peak')
        ok_count = 0
        for i in range(len(jd)):
            jd_up = jd[:i+1]; flux_up = flux[:i+1]
            k = np.argmax(flux_up)
            days_to_peak = float(jd_up[k] - jd_up[0])
            days_since_peak = float(jd_up[i] - jd_up[k]) if i!=k else 0.0
            m_to = meta_data[i, dtp_idx]
            m_si = meta_data[i, ds_idx]
            if (np.isfinite(m_to) and np.isfinite(m_si)
                and abs(m_to - days_to_peak) < 1.5
                and abs(m_si - days_since_peak) < 1.5):
                ok_count += 1
        print(f"leakage check: ok for {ok_count}/{len(jd)} events (±1.5 d)")
    else:
        print("Skipping leakage check (missing jd/logflux or peak metrics).")

    # Light curve, with spectrum time marker if present
    _plot_lc_with_spectrum_marker(event_data, event_cols, spectrum_dt=spectrum_dt if has_spec else None)

    # Show a compact meta table for the first few shown events
    T = images.shape[0]
    idxs = np.linspace(0, T-1, min(n_show, T)).astype(int).tolist()

    if len(meta_data) == T:
        meta_cols_pick = _pick_meta_columns(meta_cols)
        if meta_cols_pick:
            meta_df = pd.DataFrame(meta_data[:, [meta_cols.index(c) for c in meta_cols_pick]],
                                   columns=meta_cols_pick)
            # include event dt, band_id, and matching delta t / policy if available
            out_df = pd.DataFrame({'event_idx': np.arange(T)})
            if 'dt' in event_cols:       out_df['dt'] = event_data[:, event_cols.index('dt')]
            if 'band_id' in event_cols:  out_df['band_id'] = event_data[:, event_cols.index('band_id')].astype(int)
            out_df = pd.concat([out_df, meta_df], axis=1)

            if prov_df is not None:
                if 'alert_dt_days' in prov_df.columns:
                    out_df['alert_dt_days'] = prov_df['alert_dt_days'].astype(float).values
                if 'select_policy' in prov_df.columns:
                    out_df['select_policy'] = prov_df['select_policy'].values

            print("\nMeta snapshot for sampled events:")
            print(out_df.iloc[idxs].to_string(index=False))

    # Images for those sampled events
    for idx in idxs:
        title = f"event {idx}"
        if prov_df is not None and 'alert_dt_days' in prov_df.columns:
            title += f" | Δt={prov_df.loc[idx,'alert_dt_days']:.3f} d"
        if prov_df is not None and 'select_policy' in prov_df.columns:
            title += f" | {prov_df.loc[idx,'select_policy']}"
        show_triplet(images[idx], title=title)

    # Spectrum plot (if present)
    if has_spec and spectrum is not None:
        ttl = "Spectrum"
        if spectrum_dt is not None and np.isfinite(spectrum_dt):
            ttl += f" (dt={spectrum_dt:.3f} d)"
        if spectrum_jd is not None and np.isfinite(spectrum_jd):
            ttl += f", JD={spectrum_jd:.5f}"
        _plot_spectrum(spectrum_w if 'spectrum_wavelength' in z.files else None, spectrum, title=ttl)

    # finiteness
    print(f"images finite fraction: {np.isfinite(images).mean():.3f}")
    print("channel ranges:", [(np.nanmin(images[:,i]), np.nanmax(images[:,i])) for i in range(3)])
````

## Run visualization/validation

````python
# -------- run on a few files
cands = list_npz_files(cfg.output_root/'all', max_files=2)
if not cands:
    print("No .npz files found. Re-run build, or check DATA_DIR/SPEC_CSV.")
else:
    for p in cands:
        validate_npz(p, n_show=5)
````
